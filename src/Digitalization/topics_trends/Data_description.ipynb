```python
import pandas as pd
pd.set_option('display.max_columns', 50)

import numpy as np
import pickle
import matplotlib.pyplot as plt
from textwrap import wrap
```

```python
# upload the data
big_data = pd.read_csv("/home/seh6fy/git/publicrd/data/prd/Digital_abstract_labelled/tfidf_refined.csv")
```

```python
cb_pal = {"blue":'#377eb8', "orange":'#ff7f00', "green":'#4daf4a', "pink":'#f781bf', "brown":'#a65628', 
          "purple":'#984ea3', "gray":'#999999', "red":'#e41a1c', "yellow":'#dede00'}
```

```python
# Histogram
big_data["FY"] = big_data["FY"].astype('int')
year_counts = big_data['FY'].value_counts().sort_index(ascending=True)


year = year_counts.index.tolist()
count = year_counts.values

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_axisbelow(True)
plt.grid(True, color = "whitesmoke")
plt.bar(year, count, color='navy')
#plt.xlim(-0.7, len(year)-0.3)
plt.xlim(2007.3,2020.7)
plt.xticks(year, rotation=45)
ax.set_yticklabels(['{:,}'.format(int(x)) for x in ax.get_yticks().tolist()])
#x_ticks = ax.xaxis.get_major_ticks()
#x_ticks[-1].label1.set_visible(False)
plt.xlabel("FY")
plt.ylabel("Number of Projects")
plt.title("Data Mining & Big Data Frequency by FY")

plt.savefig("big_data_time.png", dpi = 800, bbox_inches = "tight")
```

```python
agency_counts = 100*big_data["DEPARTMENT"].value_counts()/len(big_data)

agency = agency_counts.index.tolist()
count = agency_counts.values

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_axisbelow(True)
plt.grid(True, color = "whitesmoke")
plt.bar(agency, count, color=cb_pal['blue'])
plt.ylim(0,100)
plt.xlabel("Agency")
plt.ylabel("Percent of Dataset")
plt.title("Project Distribution by Funding Agency")

plt.savefig("big_data_agency.png", dpi = 800, bbox_inches = "tight")
```
