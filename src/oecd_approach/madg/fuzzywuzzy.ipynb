```python
string_to_match = 'Mercedez-Benz'
options = ['Ford', 'Mustang', 'mersedez benz', 'MAZDA', 'Mercedez']
```

```python
process.extract(query=string_to_match, choices=options, limit=5)

# computes similarity of strings in query from seq of 
# options in choices returns a list of tuples 

# limit controls the number of tuples to return 
# each of these tuples contains two elements 
# the first one is the matching string and the 
# second one is the similarity score.

# uses default of WRatio but can change using scorer 

process.extract(query=string_to_match,
                choices=options,
                limit=3,
                scorer=fuzz.ratio)

process.extractOne(string_to_match, options)
# Returns string with highest matching score 
```
