```python
#Martha Czernuszenko's code
import pandas as pd 
import numpy as np 
import pickle 
import matplotlib.pyplot as plt 
import time 
```

```python
#Read in dataset
#df = pd.read_csv('../../src/cleaning_processing/FRAbstractsSqueakyClean.csv')
df = pd.read_pickle('../../data/working/processed_dataset_7-20.pkl')
```

```python
df.head()
```

```python
#Function to find top n words across all abstracts - need in string format
def AbstractToString(token_list, seperator=','):
    return seperator.join(token_list)
df['string_of_tokens'] = df['final_tokens'].apply(AbstractToString)
```

```python
df['string_of_tokens']
```

```python
#100
top100 = pd.Series(' '.join(df['string_of_tokens']).lower().split(",")).value_counts()[:100]
wordlist = top100.nlargest(100).index
for i in wordlist: print(i)
    
#198342 instances is least
```

```python
#150
top150 = pd.Series(' '.join(df['string_of_tokens']).lower().split(",")).value_counts()[:150]
wordlist = top150.nlargest(150).index
for i in wordlist: print(i)
```
