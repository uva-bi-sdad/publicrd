## Plot model selection results


```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
```

### LDA

```python
# coherence
c0 = pd.read_pickle("LDA/co_lda_val0.pkl")
c1 = pd.read_pickle("LDA/co_lda_val1-3.pkl")
c2 = pd.read_pickle("LDA/co_lda_val4-6.pkl")
c3 = pd.read_pickle("LDA/co_lda_val7-9.pkl")

# perplexity
p0 = pd.read_pickle("LDA/lda_p0.pkl")
p1 = pd.read_pickle("LDA/lda_p1-3.pkl")
p2 = pd.read_pickle("LDA/lda_p4-6.pkl")
p3 = pd.read_pickle("LDA/lda_p7-9.pkl")

# time
t0 = pd.read_pickle("LDA/lda_t0.pkl")
t1 = pd.read_pickle("LDA/lda_t1-3.pkl")
t2 = pd.read_pickle("LDA/lda_t4-6.pkl")
t3 = pd.read_pickle("LDA/lda_t7-9.pkl")
```

```python
# create dataframes

lda_coherence = pd.concat([c0, c1, c2, c3], axis=1)
lda_perplexity = pd.concat([p0, p1, p2, p3], axis=1)
lda_time = pd.concat([t0, t1, t2, t3], axis=1)
```

```python
lda_coherence
```

### NMF

```python
# coherence
c0 = pd.read_pickle("NMF/co_nmf_val0.pkl")
c1 = pd.read_pickle("NMF/co_nmf_val1-3.pkl")
c2 = pd.read_pickle("NMF/co_nmf_val4-6.pkl")
c3 = pd.read_pickle("NMF/co_nmf_val7-9.pkl")

# time
t0 = pd.read_pickle("NMF/nmf_t0.pkl")
t1 = pd.read_pickle("NMF/nmf_t1-3.pkl")
t2 = pd.read_pickle("NMF/nmf_t4-6.pkl")
t3 = pd.read_pickle("NMF/nmf_t7-9.pkl")
```

```python
# create dfs

nmf_coherence = pd.concat([c0, c1, c2, c3], axis=1)
nmf_time = pd.concat([t0, t1, t2, t3], axis=1)
```

```python
nmf_time
```

### Coherence plot

```python
# Add number of topics as a column for melt

lda_coherence['n_topics']=lda_coherence.index
nmf_coherence['n_topics']=nmf_coherence.index
```

```python
# change dfs from wide to long format

c_lda_melt=lda_coherence.melt(id_vars='n_topics',value_vars=['iteration {}'.format(i) for i in range(10)],value_name='Coherence' )
c_nmf_melt=nmf_coherence.melt(id_vars='n_topics',value_vars=['iteration {}'.format(i) for i in range(10)],value_name='Coherence' )

c_lda_melt['Model']='LDA'
c_nmf_melt['Model']='NMF'

all_models=pd.concat([c_lda_melt,c_nmf_melt])
```

```python
all_models
```

```python
# Coherence plots by number of topics

sns.set_style("whitegrid")
sns.lineplot(x="n_topics", y="Coherence", hue='Model', data=all_models).set(xlim=(5,50))
plt.title("LDA and NMF Coherence Comparison");
plt.xlabel('Number of Topics');
plt.ylabel('Mean $C_{V}$ Topic Coherence');

#plt.savefig("coherence_comparison.png", dpi = 800)
#plt.savefig("coherence_comparison.pdf", dpi = 800)
```

### Time plot

```python
# Add number of topics as a column for melt

lda_time['n_topics']=lda_time.index
nmf_time['n_topics']=nmf_time.index
```

```python
# change dfs from wide to long format

lda_melt=lda_time.melt(id_vars='n_topics',value_vars=['iteration {}'.format(i) for i in range(10)],value_name='Time' )
nmf_melt=nmf_time.melt(id_vars='n_topics',value_vars=['iteration {}'.format(i) for i in range(10)],value_name='Time' )

lda_melt['Model']='LDA'
nmf_melt['Model']='NMF'

all_models=pd.concat([lda_melt,nmf_melt])
```

```python
# convert to minutes

all_models["Time"] = all_models["Time"]/60
```

```python
all_models
```

```python
# Time plots by number of topics

sns.set_style("whitegrid")
sns.lineplot(x="n_topics", y="Time", hue='Model', data=all_models).set(xlim=(5,50))
plt.title("LDA and NMF Time Comparison");
plt.xlabel('Number of Topics');
plt.ylabel('Mean Time (minutes)');

#plt.savefig("time_comparison.png", dpi = 800)
#plt.savefig("time_comparison.pdf", dpi = 800)
```
