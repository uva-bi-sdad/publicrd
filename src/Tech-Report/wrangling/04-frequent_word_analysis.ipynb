```python
#Martha Czernuszenko's code
import pandas as pd 
import numpy as np 
import pickle 
import matplotlib.pyplot as plt 
import time 
```

```python
#Read in dataset

df = pd.read_pickle('../../../data/prd/Tech-Report/FR_processed_21SEPT14.pkl')
```

```python
df.head()
```

```python
#Function to find top n words across all abstracts - need in string format
def AbstractToString(token_list, seperator=','):
    return seperator.join(token_list)
df['string_of_tokens'] = df['final_tokens'].apply(AbstractToString)
```

```python
df['string_of_tokens']
```

```python
wrd_frq = pd.Series(','.join(df['string_of_tokens']).split(",")).value_counts()[:300]
```

```python
wrd_frq[0:40]
```

```python
# top words are generally the same as they were in the old dataset

num_wds = 200

top_wds = wrd_frq.nlargest(num_wds).index
for i in top_wds: print(i)
```
