```python
stopWords = set(nltk.corpus.stopwords.words('english'))
ps=PorterStemmer()
```

```python
#Creates a series object of preprocessed words (lower case, minimal length 3, excluding stop words, then stemmed using PorterStemmer)
my_list=pd.Series([ps.stem(x) for x in simple_preprocess(long_str, deacc=True,min_len=3) if not x in stopWords])
#Count the frequency of the stemmed words and write to file 
my_list.value_counts().to_csv('Test.csv')
```
