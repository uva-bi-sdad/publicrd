```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
```

# LDA

```python
c0 = pd.read_pickle("lda_c0-2.pkl")
c1 = pd.read_pickle("lda_c3-5.pkl")
c2 = pd.read_pickle("lda_c6-7.pkl")
c3 = pd.read_pickle("lda_c8-9.pkl")

p0 = pd.read_pickle("lda_p0-2.pkl")
p1 = pd.read_pickle("lda_p3-5.pkl")
p2 = pd.read_pickle("lda_p6-7.pkl")
p3 = pd.read_pickle("lda_p8-9.pkl")
```

```python
lda_coherence = pd.concat([c0, c1, c2, c3], axis=1)
```

```python
#lda_coherence["average"] = np.mean(lda_coherence, axis=1)
```

```python
lda_coherence
```

```python
lda_perplexity = pd.concat([p0, p1, p2, p3], axis=1)
```

```python
lda_perplexity["average"] = np.mean(lda_perplexity, axis=1)
```

```python
lda_perplexity
```

```python
# plot results

num_topics = lda_coherence.index
avg_c = lda_coherence["average"]
avg_p = lda_perplexity["average"]

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(num_topics, avg_c, marker='o')
plt.title("LDA")
plt.xlabel("Number of Topics")
plt.ylabel("$C_{V}$ Topic Coherence")
#plt.legend(("coherence"), loc='best')

plt.subplot(1,2,2)
plt.plot(num_topics, avg_p, marker='o')
plt.title("LDA")
plt.xlabel("Number of Topics")
plt.ylabel("Perplexity")
#plt.legend(("perplexity"), loc='best')

plt.subplots_adjust(wspace=0.3)
#plt.tight_layout()
plt.show()
```

# NMF

```python
c0 = pd.read_pickle("nmf_c0-1.pkl")
c1 = pd.read_pickle("nmf_c2-3.pkl")
c2 = pd.read_pickle("nmf_c4-5.pkl")
c3 = pd.read_pickle("nmf_c6-7.pkl")
c4 = pd.read_pickle("nmf_c8-9.pkl")
```

```python
nmf_coherence = pd.concat([c0, c1, c2, c3, c4], axis=1)
```

```python
nmf_coherence
```

# Create Seaborn plot

```python
# Add number of topics as a column for melt

lda_coherence['n_topics']=lda_coherence.index
nmf_coherence['n_topics']=nmf_coherence.index
```

```python
# change dfs from wide to long format

c_lda_melt=lda_coherence.melt(id_vars='n_topics',value_vars=['iteration {}'.format(i) for i in range(10)],value_name='Coherence' )
c_nmf_melt=nmf_coherence.melt(id_vars='n_topics',value_vars=['iteration {}'.format(i) for i in range(10)],value_name='Coherence' )

c_lda_melt['Model_Type']='LDA'
c_nmf_melt['Model_Type']='NMF'

all_models=pd.concat([c_lda_melt,c_nmf_melt])
```

```python
# Coherence plots by number of topics

sns.set_style("whitegrid")
sns.lineplot(x="n_topics", y="Coherence", hue='Model_Type', data=all_models)
plt.xlabel('Number of Topics');
```
